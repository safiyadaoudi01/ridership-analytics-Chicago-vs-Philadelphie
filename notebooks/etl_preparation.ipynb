## Import et préparation des données

### Fusionnement des fichiers csv de chicago

```python
import pandas as pd

df_chicago_daily = pd.read_csv('..\data\chicago_csv\chicago_ridership_daily.csv')
df_chicago_monthly=pd.read_csv('..\data\chicago_csv\chicago_ridership_monthly.csv')
```

```python
df_chicago_daily.info()
```

```python
df_chicago_monthly.head()
```

```python
df_chicago_monthly.info()
```

```python
import pandas as pd

# Extraire l'année et le mois de df_chicago_daily
df_chicago_daily['date'] = pd.to_datetime(df_chicago_daily['date'])
df_chicago_daily['year_month'] = df_chicago_daily['date'].dt.to_period('M')

# Transformer month_beginning en année-mois
df_chicago_monthly['month_beginning'] = pd.to_datetime(df_chicago_monthly['month_beginning'])
df_chicago_monthly['year_month'] = df_chicago_monthly['month_beginning'].dt.to_period('M')

# Fusionner sur route et year_month
df_chicago = pd.merge(
    df_chicago_daily,
    df_chicago_monthly,
    how='left',
    left_on=['route', 'year_month'],
    right_on=['route', 'year_month']
)

# Vérifier le résultat
df_chicago.head()
```

### Appel d'api de CTA Bus Tracker (Chicago)

```python
import requests
import pandas as pd

CTA_API_KEY = "VOTRE_CLE_API"  # Remplacez par votre clé
CTA_BASE_URL = "https://www.ctabustracker.com/bustime/api/v3"

def  get_cta_routes(api_key):
    params = {"key": api_key, "format": "json"}
    response = requests.get(f"{CTA_BASE_URL}/getroutes", params=params)
    data = response.json().get("bustime-response", {}).get("routes", [])
    return pd.DataFrame(data)
```
