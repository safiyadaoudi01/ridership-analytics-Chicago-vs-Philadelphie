## EDA (Exploratory Data Analysis)

### Route

```python
df_phil_route.head()
```

```python
df_chicago_route.head()
```

#### Ajouter la colonne 'month' dans 'df_chicago_route' pour faire une comparaison mensuelle

```python
df_chicago_route["month"]=df_chicago_route["date"].dt.month
```

#### Suppression des colonnes inutile dans df_phil_route

```python
df_phil_route=df_phil_route.drop(columns=['Source', 'ObjectId'])
```

#### Grouper par l'année,le mois et la route pour avoir la moyenne par mois de 'ridership' dans df_chicago_route

```python
df_chicago_route_monthly = (
    df_chicago_route
    .groupby(['year', 'month', 'route'], as_index=False)
    .agg({
        'ridership': 'mean'
    })
)
```

### Vérification des colonnes

```python
df_chicago_route_monthly.columns
```

```python
df_phil_route.columns
```

#### Fusionnement de df_chicago_route_monthly et df_phil_route

```python
df_chicago_route_clean = df_chicago_route_monthly.rename(columns={
    'ridership': 'avg_ridership_chicago'
})
```

```python
df_phil_route_clean = df_phil_route.rename(columns={
    'Calendar_Year': 'year',
    'Calendar_Month': 'month',
    'Route': 'route',
    'Average_Daily_Ridership': 'avg_ridership_philadelphia'
})
```

```python
df_route_comparison = pd.merge(
    df_chicago_route_clean,
    df_phil_route_clean,
    on=['year', 'month', 'route'],
    how='inner'
)
```

```python
df_route_comparison.head()
```

```python
df_route_comparison.info()
```

```python
df_route_comparison.to_csv('../data/processed/route_comparison.csv', index=False)
```

#### les modes qu'on va comparer concerne juste 'bus' et 'rail'

```python
df_phil_mode=df_phil_mode[df_phil_mode['Mode'].isin(['Bus', 'Heavy Rail', 'Regional Rail'])]
```

```python
df_phil_mode['Mode'].unique()
```

#### Ajout de la colonne mois dans df_chicago_mode pour faire une comparaison mensuelle

```python
df_chicago_mode["month"]=df_chicago_mode["service_date"].dt.month
```

```python
df_chicago_mode.head()
```

#### Restructurer les colonnes de df_chicago_mode pour etre comparable avec df_phil_mode

##### Étape 1 : agrégation mensuelle (moyenne journalière)

```python
df_chicago_monthly_mode = (
    df_chicago_mode
    .groupby(['year', 'month'], as_index=False)
    .agg({
        'bus': 'mean',
        'rail_boardings': 'mean'
    })
)
```

```python
df_chicago_monthly_mode.head()
```

##### Étape 2 : passage en format long (mode)

```python
df_chicago_monthly_mode = df_chicago_monthly_mode.melt(
    id_vars=['year', 'month'],
    value_vars=['bus', 'rail_boardings'],
    var_name='mode',
    value_name='ridership'
)
```

```python
df_chicago_monthly_mode.head()
```

#### Renommer les variables 'Heavy Rail' et 'Regional Rail' comme 'rail_boardings' dans df_phil_mode pour le comparer avec 'rail_boardings' de df_chicago_monthly_mode

```python
df_phil_mode['Mode'] = df_phil_mode['Mode'].replace({
    'Heavy Rail': 'rail_boardings',
    'Regional Rail': 'rail_boardings'
})
```

```python
df_phil_mode['Mode'] = df_phil_mode['Mode'].replace({'Bus': 'bus'})
```

```python
df_phil_mode['Mode'].unique()
```

#### Supprimer les dubliqués dans df_phil_mode aprer renommer les deux types de Rail juste par 'rail_boardings'

```python
df_phil_mode.duplicated(subset=['Calendar_Year', 'Calendar_Month', 'Mode']).sum()
```

```python
df_phil_mode=df_phil_mode.groupby(['Calendar_Year', 'Calendar_Month','Mode'], as_index=False).agg({
        'Average_Daily_Ridership': 'mean'})
```

```python
df_phil_mode.duplicated(subset=['Calendar_Year', 'Calendar_Month', 'Mode']).sum()
```

#### Fusionnement de df_chicago_monthly_mode et df_phil_mode

```python
df_chicago_mode_clean = df_chicago_monthly_mode.rename(columns={
    'ridership': 'avg_ridership_chicago'
})
```
